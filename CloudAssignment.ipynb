# Some Visulization


#### Histogram Distribution of Numerical Variables in Dataset

#### ratings change over the year

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='original_publication_year', y='average_rating', data=dataset)
plt.xticks(rotation=90)
plt.title('Average Rating by Publication Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

# Most Selling Book

```python
most_selling=harrypotter.sort_values(by='books_count',ascending = False )
most_selling_book = most_selling.iloc[0]
print("The most selling Harry Potter book based on book count is:")
print(most_selling_book[['title', 'books_count']])
```

We could also sort all of them then print it and visualize it

```python
print(most_selling[['title', 'books_count']])
```

```python
plt.bar(most_selling['title'], most_selling['books_count'], color='blue')  # Create bar plot
plt.xlabel('Title')  
plt.ylabel('Books Count') 
plt.title('Harry Potter Books by Books Count')  
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()  
```
